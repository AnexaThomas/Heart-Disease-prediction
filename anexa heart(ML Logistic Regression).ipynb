```python
#importing modules
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score #to check how our well are our model
```

```python
#data collection and processing
#loading csv files into a pandas Dataframe
heart_data=pd.read_csv("user_data.csv")
heart_data
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
heart_data["Gender"]=le.fit_transform(heart_data["Gender"])

heart_data
```

```python
#finding the desctribution of the target value
heart_data["Purchased"].value_counts() #1 is represending who have heart disease and 0 represending who don't have heart disease
```

```python
#splitting the features and target
x=heart_data.drop(columns="Purchased",axis=1)  # it will make the 2 tables 1 st one is containg all features
y=heart_data["Purchased"]  # and 2 nd table contaning target value
```

```python
x
```

```python
y
```

```python
#splitting data into training data and test data
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,shuffle=False )
                                               #Shuffle=False means it willnot shuffle the value
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

```python
#Model Training logistic Regression
model=LogisticRegression()
```

```python
#training the logistic Regression model with training data
model.fit(x_train,y_train)
```

```python
#model Evaluation
#accuracy score
x_train_prediction=model.predict(x_train)
training_data_accuracy=accuracy_score(x_train_prediction,y_train)
training_data_accuracy.round(2)
```

```python
print("Accuracy on Training data:",training_data_accuracy)
```

```python
x_test_prediction=model.predict(x_test)
test_data_accuracy=accuracy_score(x_test_prediction,y_test)
test_data_accuracy.round(2)
```

```python
print("Accuracy on Test data:",test_data_accuracy)
```
